Check Percentage of Train, Validation and Test Data

```python
from pathlib import Path

import pandas as pd
from sklearn.model_selection import train_test_split

ROOT_DIR = Path.cwd()
if not (ROOT_DIR / "data").exists():
    ROOT_DIR = ROOT_DIR.parent

RAW_DATA_PATH = ROOT_DIR / "data" / "raw" / "emails.csv"
PROCESSED_DATA_DIR = ROOT_DIR / "data" / "processed"

df = pd.read_csv(RAW_DATA_PATH)

# First split: 70% train, 30% temp
train_df, temp_df = train_test_split(
    df,
    test_size=0.30,
    random_state=42,
    stratify=df["spam"]
)

# Second split: temp 30% becomes 20% validation and 10% test
# validation = 20/30 = 0.6667 of temp
# test = 10/30 = 0.3333 of temp
validation_df, test_df = train_test_split(
    temp_df,
    test_size=1/3,
    random_state=42,
    stratify=temp_df["spam"]
)

print("Train shape:", train_df.shape)
print("Validation shape:", validation_df.shape)
print("Test shape:", test_df.shape)
```

Create Raw data to Processed Data(Train, Validation, Test)


```python
PROCESSED_DATA_DIR.mkdir(parents=True, exist_ok=True)

train_df.to_csv(PROCESSED_DATA_DIR / "train.csv", index=False)
validation_df.to_csv(PROCESSED_DATA_DIR / "validation.csv", index=False)
test_df.to_csv(PROCESSED_DATA_DIR / "test.csv", index=False)
```

Check Percentage

```python
total = len(df)

print("Train %:", len(train_df) / total * 100)
print("Validation %:", len(validation_df) / total * 100)
print("Test %:", len(test_df) / total * 100)
```
