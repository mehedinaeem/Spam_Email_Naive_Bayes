```python
df["label"] = df["spam"].map({
    0: "Not Spam",
    1: "Spam"
})

sns.countplot(x="label", data=df, order=["Not Spam", "Spam"])
plt.title("Spam vs Not Spam Emails")
plt.xlabel("Email Type")
plt.ylabel("Count")
plt.show()
```
